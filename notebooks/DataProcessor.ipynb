```python
import pickle

with open("../data/match_data.pkl", "rb") as f:
    matchData = pickle.load(f)

len(matchData)
```

```python
from sklearn.model_selection import train_test_split
len(matchData)*0.1129

trainData, valData = train_test_split(matchData, test_size=0.1129, random_state=42)
```

```python
len(valData)
len(trainData)
print(len(trainData), len(valData))

trainData2, testData = train_test_split(trainData, test_size=0.1273, random_state=42)
print(len(trainData2), len(testData))
```

```python
import cassiopeia.data as data
def GetSideString(teamOrPlayer):
    if teamOrPlayer["side"] == data.Side.blue:
        return "blue"
    return "red"

for i,match in enumerate(valData):
  for j,team in enumerate(match["teams"]):
    valData[i]["teams"][j]["side"] = GetSideString(team)
    for y, player in enumerate(team["participants"]):
      valData[i]["teams"][j]["participants"][y]["side"] = GetSideString(player)
```

```python
import json

matchData = {
  "matches": valData
}

with open("../data/val_data.json", "w") as f:
  f.write(json.dumps(matchData))
```
